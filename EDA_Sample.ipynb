# This is a sample EDA project

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

Importing data

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.dtypes
```

```python
df = df.drop(['Popularity', 'Vehicle Style'], axis=1)
df.head()
```

```python
df = df.rename(columns={'Engine HP':'HP', 'Engine Cylinders':'Cylinders', 'Number of Doors':'Doors', 'Market Category':'Category', "highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price"})
df.head()
```

```python
df = df.rename(columns={'Transmission Type':'Transmission', 'Engine Fuel Type':'Fuel', 'Vehicle Size':'Size'})
df.head()
```

```python
df.shape
```

```python
print("Number of duplicate rows: ", df[df.duplicated()].shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df['HP'] = df.groupby('Model')['HP'].transform(lambda x: x.fillna(x.mean()))
```

```python
print(df.isnull().sum())
```

```python
df['HP'].fillna(df['HP'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Transmission'].unique()
```

```python
df.isnull().sum()
```

```python
df['Fuel'].unique()
```

```python
df.loc[df['Fuel'] == 'electric', 'Cylinders'] = df.loc[df['Fuel'] == 'electric', 'Cylinders'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['Cylinders'] = df.groupby('Make')['Cylinders'].transform(lambda x: x.fillna(x.mode()))
df.isnull().sum()
```

```python
df.loc[df['Model'] == 'RX-7', 'Cylinders'] = df.loc[df['Model'] == 'RX-7', 'Cylinders'].fillna(2)
df.loc[df['Model'] == 'RX-8', 'Cylinders'] = df.loc[df['Model'] == 'RX-8', 'Cylinders'].fillna(4)
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Category'], axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Fuel'] = df['Fuel'].fillna(df['Fuel'].mode()[0])
df['Doors'].fillna(df['Doors'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

Outliers Detection

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df['HP'].quantile(0.25)
Q3 = df['HP'].quantile(0.75)
IQR_HP = Q3 - Q1
print(IQR_HP)
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR_Price = Q3 - Q1
print(IQR_Price)
```

```python
df['Price'] = df['Price'][~((df['Price'] < (Q1 - 1.5 * IQR_Price)) | (df['Price'] > (Q3 + 1.5 * IQR_Price)))]
```

```python
sns.boxplot(x=df['Price'])
```
