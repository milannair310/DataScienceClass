```python
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df.dropna()  #drop all null or empty columns
```

```python
df.drop('Survived',axis = 1,inplace = True)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True) #try doing Name 'stead of Age, its categorical data so mean won't work
df.isnull().sum()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])#note that datetime is just converting a string into a datetime type, you can do the same thing with int or float
```

```python
arr1 = [1,2,3]
arr2 = [2,3,4]
arr3 = [5,6,7]
arr = np.stack((arr1, arr2, arr3), axis = 1) #takes columns
print(arr)
arr5 = np.hstack((arr1, arr2, arr3)) #does the same thing as concatinate
print(arr5)
arr9 = np.dstack((arr1, arr2, arr3)) #stacks rows in columns and columns in rows
print(arr9)
x = np.where(arr == 3) #finds where arr==4
print(x)
```
